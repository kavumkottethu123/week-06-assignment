```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Read the data from the csv file
students_data= pd.read_csv('StudentsPerformance.csv')
```

```python
students_data
```

## 1. Find out how many males and females participated in the test.


```python
#Counting the no:of males and females from the gender column, dropna=False keyword argument to also count missing values:
students_data["gender"].value_counts(dropna=False)
```

```python
# grouping by ‘race/ethnicity’ and ‘gender’. This should return the number of data of each race segregated by gender.
students_data.groupby(['gender', 'race/ethnicity']).size()
```

## 2. What do you think about the students' parental level of education'?


```python
#Education of parents
students_data['parental level of education'].unique()
```

```python
#counting the education qualification level
students_data['parental level of education'].value_counts()
```

```python
 #grouping by ‘race/ethnicity’ ,‘gender’and'parental level of education'.
students_data.groupby(['gender', 'race/ethnicity','parental level of education']).size()
```

## 3. Who scores the most on average for math, reading and writing based on
● Gender
● Test preparation course

```python
#scores the most on average for math, reading and writing.
# round function is to rounds up the floating-point numbers to the desired number of decimal points.
score_read=students_data[['gender','test preparation course','math score','reading score','writing score']].groupby(['gender','test preparation course']).agg(['mean','max','count']).round(1)
score_read.columns=['max math score', 'average math score','count math sore' ,'max reading Score', 'average reading score', 'count read score','max writing score','average writing score','count writing score']
```

```python
#scores based on gender and test preparation course.
score_read
```

## 4. What do you think about the scoring variation for math, reading and writing
based on
● Gender
● Test preparation course

```python
# scoring variation for math, reading and writing
students_data[['gender','test preparation course','math score','reading score','writing score']].groupby(['gender','test preparation course']).agg(['mean','var','count']).round(1)
```
